```python
df = pd.read_json(DATA_DIR / "pcos_without_infertility.jsonl", lines=True)
```

```python
df.head()
```

```python
df = df.copy()
df.head()
```

```python
def plot_column(dataframe, column_name=None, xlabel="x", ylabel="Y", title="Title", kind="bar"):
    if column_name is not None:
        dataframe[column_name].value_counts().plot(kind=kind)
    else:
        dataframe.value_counts().plot(kind=kind)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
columns_to_drop = ["Sl. No", "Patient File No."]
```

```python
df = df.drop(columns_to_drop, axis=1)
df.head()
```

```python
labels = df["PCOS (Y/N)"]
labels.head()
```

```python
plot_column(dataframe=labels, xlabel="Y/N", ylabel="Size of Data", title="PCOS Data") 
```

```python
df = df.drop(["PCOS (Y/N)"], axis=1)
df.head()
```

```python
plot_column(dataframe=df, column_name="Follicle No. (R)_flag", kind="bar")
```
